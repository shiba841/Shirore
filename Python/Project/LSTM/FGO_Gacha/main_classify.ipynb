```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def _load_data(data, seq_len=100):
    data = data[0:-(len(data) % (seq_len+1))]
    data_re = data.reshape(-1, seq_len+1)
    x = data_re[:, 0:seq_len]
    x = np.expand_dims(x, -1)
    y = data_re[:, -1]
        
    return np.array(x), np.array(y)
```

```python
raw_csv = pd.read_csv("./fgo_gacha_raw.csv", header=None).dropna().astype(int)
raw_arr = np.array(raw_csv).reshape(1, -1)
data_arr = raw_arr[raw_arr <= 33]
data_arr = np.where(data_arr == 11, 10, data_arr)
data_arr = np.where(data_arr == 22, 11, data_arr)
data_arr = np.where(data_arr == 33, 12, data_arr)
data_arr = data_arr - 1
```

```python
# 1系列データの長さ
length_of_sequence = 10
# 全データに対するテストデータの割合
train_data_ratio = 0.9
train_size = int(round(len(data_arr) * train_data_ratio))

X_train, y_train = _load_data(data_arr[0:train_size], length_of_sequence)
X_test, y_test = _load_data(data_arr[train_size:], length_of_sequence)
```

```python
y_test.shape
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.layers.recurrent import LSTM
```

```python
labels = 12
in_out_neurons = 1
hidden_neurons = 100

model = Sequential()
model.add(LSTM(hidden_neurons,
               batch_input_shape=(None, length_of_sequence, 1),
               return_sequences=False))
model.add(Dense(labels))
model.add(Activation("softmax"))
model.compile(loss="sparse_categorical_crossentropy", optimizer="rmsprop", metrics=['accuracy'])
```

```python
summary = model.fit(X_train, y_train, batch_size=600, epochs=1000, validation_split=0.05)
```

```python
plt.plot(summary.history['acc'])
```

```python
predicted = model.predict(X_test)
predicted.shape
```

```python
show_data = 50
predict_label = np.argmax(predicted[:show_data], axis=1)
result = pd.DataFrame(predict_label, columns=["predict"])
result["true"] = y_test[:show_data]
#result
```
